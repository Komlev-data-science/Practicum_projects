````python
# Рассчет данных по выручке
revenue = pa_diploma_orders.groupby('order_month').agg({'order_id':'nunique',
                                                        'total_price':'sum'}).reset_index()
revenue['revenue'] = revenue['total_price']*0.05
````

````python
# Визуализация выручки
plt.figure(figsize=(12,5))
plt.plot(revenue['order_month'], revenue['revenue'], marker="o")
plt.title('Выручка')
plt.xlabel('Даты')
plt.ylabel('Значения, млн. у.е.')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

````python
# Рассчет DAU
dau = pa_diploma_events.groupby(pa_diploma_events['event_date'].dt.to_period('D'))['user_id'].nunique().reset_index()
dau['event_date'] = dau['event_date'].dt.to_timestamp()
# Визуализация DAU
plt.figure(figsize=(12,5))
plt.plot(dau['event_date'],dau['user_id'])
plt.title('DAU')
plt.xlabel('Даты')
plt.ylabel('Значения')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

````python
# Рассчет MAU
mau = pa_diploma_events.groupby(pa_diploma_events['event_date'].dt.to_period('M'))['user_id'].nunique().reset_index()
mau['event_date'] = mau['event_date'].dt.to_timestamp()

# Визуализация MAU
plt.figure(figsize=(12,5))
plt.plot(mau['event_date'],mau['user_id'], marker='o')
plt.title('MAU')
plt.xlabel('Даты')
plt.ylabel('Значения')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

````python
# Рассчет stikiness
pa_diploma_events['event_date'] = pd.to_datetime(pa_diploma_events['event_date'])
pa_diploma_events['date'] = pa_diploma_events['event_date'].dt.date
pa_diploma_events['month'] = pa_diploma_events['event_date'].dt.to_period('M').dt.to_timestamp()
dau = pa_diploma_events.groupby('date')['user_id'].nunique().reset_index()
dau['date'] = pd.to_datetime(dau['date'])
dau['month'] = dau['date'].dt.to_period('M').dt.to_timestamp()
dau_monthly_average = dau.groupby('month')['user_id'].mean().reset_index()
mau = pa_diploma_events.groupby('month')['user_id'].nunique().reset_index()
stickiness_table = pd.merge(mau, dau_monthly_average, how='left', on='month')
stickiness_table['stick'] = 100*(stickiness_table['user_id_y']/stickiness_table['user_id_x'])

# Визуализация Stikiness
plt.figure(figsize=(15,5))
plt.plot(stickiness_table['month'], stickiness_table['stick'], marker='o')
plt.title('Stickiness')
plt.xlabel('Месяцы')
plt.ylabel('Значения %')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

````python
# Рассчет MAC
pa_diploma_orders['order_day'] = pa_diploma_orders['order_date'].dt.to_period('D')
maс = pa_diploma_orders.groupby('order_month', as_index = False).agg(
                        mac=('user_id','nunique'))
````

````python
# Визуализация MAC
plt.figure(figsize=(12,5))
sns.lineplot(data = maс, x = 'order_month', y = 'mac', marker='o')
plt.grid()
plt.xlabel('Месяц')
plt.title('MAС за весь период')
plt.xticks(rotation=45)
plt.show()
````

````python
# Рассчет CR
CR = pd.merge(maс,mau, left_on='order_month',right_on='month')
mac = pd.to_datetime(maс['order_month'])
CR['CR'] = 100*(CR['mac']/CR['user_id'])
````

````python
# Визуализация CR
CR['month'] = CR['month'].astype('str')
plt.figure(figsize=(12,5))
sns.lineplot(data = CR, x = 'month', y = 'CR', marker='o')
plt.grid()
plt.xlabel('Месяц')
plt.ylabel('Конверсия в покупку в %')
plt.title('CR')
plt.xticks(rotation=45)
plt.show()
````

### 3. Оценка метрик монетизации и юнит-экономики

````python
# Рассчет arpu
arpu = pd.merge(revenue[['order_month','revenue']], mau, left_on='order_month', right_on='month')
arpu['arpu'] = arpu['revenue']/arpu['user_id']
````

````python
# Визуализация ARPU
plt.figure(figsize=(12,5))
plt.plot(arpu['month'], arpu['arpu'], marker='o')
plt.title('ARPU')
plt.xlabel('Месяцы')
plt.ylabel('Средний доход на пользователя, у.е.')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

````python
# Рассчет ARPPU
paying_users = pa_diploma_orders.groupby('order_month')['user_id'].nunique().reset_index(name='pu')
revenue = revenue.merge(paying_users, on='order_month')
revenue['arppu'] = revenue['revenue'] / revenue['pu']
````

````python
# Визуализация ARPPU
plt.figure(figsize=(12,5))
plt.plot(revenue['order_month'], revenue['arppu'], marker='o')
plt.title('ARPPU')
plt.xlabel('Месяцы')
plt.ylabel('Средний доход на платящего пользователя, у.е.')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

````python
# Визуализация AOV
plt.figure(figsize=(12,5))
revenue['aov'] = revenue['revenue'] / revenue['order_id']
plt.plot(revenue['order_month'], revenue['aov'], marker='o')
plt.title('AOV')
plt.xlabel('Месяцы')
plt.ylabel('Средняя стоимость заказа, у.е.')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

````python
# Рассчет CAC
pa_diploma_users_merged = pd.merge(pa_diploma_campaign_costs, campaing_costs_grouped, left_on=['acq_channel','spend_month'], right_on=['acq_channel','cohort_month'])
pa_diploma_users_merged['CAC'] = pa_diploma_users_merged['budget'] / pa_diploma_users_merged['user_id']
CAC = pa_diploma_users_merged.groupby('cohort_month')['CAC'].mean().reset_index()
````

````python
# Визуализация CAC
plt.figure(figsize=(12,5))
plt.plot(CAC['cohort_month'], CAC['CAC'], marker='o')
plt.title('CAC')
plt.xlabel('Месяцы')
plt.ylabel('Cтоимость привлечения, у.е.')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

````python
# Группируем по месяцам регистрации и типу трафика, считая уникальных пользователей
registrations = user_orders_df.groupby(['cohort_month', 'acq_channel'])['user_id'].nunique().reset_index()
# Переименовываем столбец с количеством уникальных пользователей
registrations = registrations.rename(columns={'user_id': 'registrations'})

# Группируем пользователей, которые заплатили, по каналу и месяцу
paid_users = pd.merge(pa_diploma_orders,pa_diploma_users, on='user_id', how='inner').drop_duplicates(subset='user_id').groupby(['order_month', 'acq_channel'])['user_id'].nunique().reset_index()
# Переименовываем столбец с количеством уникальных пользователей
paid_users = paid_users.rename(columns={'user_id': 'paid_users'})

# Объединяем таблицы по месяцам и каналу
conversion_data = pd.merge(
    registrations,
    paid_users,
    left_on=['cohort_month', 'acq_channel'],
    right_on=['order_month', 'acq_channel'],
    how='left'
)
# Заменяем пропущенные значения нулями
conversion_data['paid_users'] = conversion_data['paid_users'].fillna(0)

# Рассчитываем конверсию
conversion_data['conversion_rate'] = conversion_data['paid_users'] / conversion_data['registrations']

# Строим график конверсии по каналам и месяцу регистрации
plt.figure(figsize=(12, 6))
# Пишем цикл по каналам привлечения
for acq_channel in conversion_data['acq_channel'].unique():
    # Фильтруем данные по типу трафика
    traffic_data = conversion_data[conversion_data['acq_channel'] == acq_channel]
    # Сортируем по месяцу для правильного порядка
    traffic_data = traffic_data.sort_values('cohort_month')
    # Строим линию для каждого типа трафика
    plt.plot(traffic_data['cohort_month'], traffic_data['conversion_rate'], label=acq_channel)

# Настроим внешний вид графика
plt.title('CR rate по каналам')
plt.xlabel('Month')
plt.ylabel('CR rate')
plt.xticks(rotation=45)
plt.legend(title='Traffic Type', loc='lower left')
plt.grid(True)
plt.show()
````

````python
CR = user_orders_df.groupby(['acq_channel','user_id','cohort_month'])['revenue'].sum().reset_index()

# 1. Создаем признак покупки
CR['purchase'] = CR['revenue'].apply(lambda x: 1 if x > 0 else 0)

# 2. Рассчитываем размеры когорт
cohort_sizes = CR.groupby(['cohort_month', 'acq_channel'])['user_id'].nunique().reset_index(name='cohort_size')

# 3. Считаем количество покупателей в каждой когорте
buyers_counts = CR.groupby(['cohort_month', 'acq_channel'])['purchase'].sum().reset_index(name='buyers')

# 4. Объединяем данные и рассчитываем конверсию
conversion = cohort_sizes.merge(buyers_counts, on=['cohort_month', 'acq_channel'], how='left')
conversion['conversion_rate'] = conversion['buyers'] / conversion['cohort_size']

# 5. Преобразуем даты в строки для удобства отображения
conversion['cohort_month_str'] = conversion['cohort_month'].dt.strftime('%Y-%m-%d')

# 6. Создаем сводную таблицу для heatmap
pivot_table = conversion.pivot_table(
    index='cohort_month_str',
    columns='acq_channel',
    values='conversion_rate',
    fill_value=0  # Заполняем нулями отсутствующие значения
)

# 7. Визуализация
plt.figure(figsize=(12, 10))
sns.heatmap(
    pivot_table,
    annot=True,
    fmt=".2%",
    cmap='coolwarm',
    linewidths=.5,
    cbar_kws={'label': 'Конверсия в покупку'}
)
plt.title('Конверсия в покупку по месяцу привлечения когорты и каналам привлечения')
plt.xlabel('Канал привлечения')
plt.ylabel('Месяц привлечения')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
````

## 1. Контекст и цель эксперимента

- **Бизнес-контекст:** Команда продукта запускает эксперимент, направленный на повышение конверсии среди пользователей, привлечённых через канал TikTok. Для стимулирования первой покупки пользователям будет предоставлен бонус на первую покупку.  
- **Цель:** Проверить, влияет ли предоставление бонуса на первую покупку на конверсию и выручку среди пользователей TikTok.  
- **Гипотеза:** Предоставление бонуса на первую покупку новым пользователям, привлеченным через TikTok, приведет к увеличению конверсии в первую покупку

---

## 2. Формат эксперимента

- **Тип эксперимента:** Сплит-роллинг (rolling split)  
- **Канал:** Только пользователи, пришедшие с TikTok  
- **Группы:**  
  - **Контрольная (A):** пользователи TikTok без бонуса  
  - **Тестовая (B):** пользователи TikTok с бонусом на первую покупку  
- **Распределение:** Равномерное распределение новых пользователей TikTok между группами по дате регистрации  
- **Период проведения:** 3 недели (фиксировано)

---

## 3. Ключевые метрики

### Основная метрика

- **Конверсия в первую покупку** — доля пользователей, зарегистрировавшихся с TikTok и совершивших первую покупку  
  - *Цель:* увеличить конверсию с помощью бонуса  
  - *Критерий успеха:* статистически значимый рост конверсии в тестовой группе

### Вспомогательные метрики

- **ARPU** (средняя выручка на пользователя TikTok)  
- **ARPPU** (средняя выручка на покупателя TikTok)  
- **AOV** (средний чек TikTok)

### Барьерная метрика

- **Общая выручка от пользователей TikTok**

---

## 4. Размеры групп и запуск

- **Целевая аудитория:** только пользователи TikTok  
- **Размеры групп:** определяются по фактическому трафику TikTok за период эксперимента

````python
alpha = 0.05 ## на каком уровне значимости проверяем гипотезу о равенстве вероятностей
m_a = split_share[split_share['group']=='control']['user_id'].values
m_b = split_share[split_share['group']=='treatment']['user_id'].values
n_a = m_a+m_b
n_b = m_a+m_b
stat_ztest, p_value_ztest = proportions_ztest(
    [m_a, m_b],
    [n_a, n_b],
    alternative='two-sided' # так как H_1: p_a != p_b
)
p_value_ztest
# 0.43026503

if p_value_ztest > alpha:
    print(f'pvalue={p_value_ztest} > {alpha}')
    print('Нулевая гипотеза находит подтверждение! Разница между группами 0.6 % является допустимой и не имеет статистически значимой разницы')
else:
    print(f'pvalue={p_value_ztest} < {alpha}')
    print('Нулевая гипотеза не находит подтверждения!')
````

Проверяем гипотезу: Нулевая гипотеза H0: Доля переходящих от регистрации к первому заказу не отличается статистически между группами control и treatment.

Альтернативная гипотеза H1: Доля переходящих от регистрации к первому заказу показывает статистически значимый рост от группы control к группе treatment.

````python
# Получаем список всех уникальных групп
groups = sorted(kpi_metric['group'].unique())

# Создаём словарь для хранения данных по группам
group_data = {group: kpi_metric[kpi_metric['group'] == group]['CR'] for group in groups}
first_data = {group: kpi_metric[kpi_metric['group'] == group]['first_paid'] for group in groups}
reg_data = {group: kpi_metric[kpi_metric['group'] == group]['registrations'] for group in groups}

# Создаём список для хранения результатов
results = []

# Проводим попарные сравнения всех групп
for i in range(len(groups)):
   for j in range(i + 1, len(groups)):
       group1 = groups[i]
       group2 = groups[j]
       # Извлекаем данные для групп
       data1 = first_data[group1]
       data2 = first_data[group2]
       data3 = reg_data[group1]
       data4 = reg_data[group2]

        # Вычисляем необходимые статистики для z-теста
       count = [data1.sum(), data2.sum()]  # Количество успехов
       nobs = [data3.sum(), data4.sum()]     # Размеры выборок
       _, p_value = proportions_ztest(count, nobs, alternative='smaller')

       # Добавляем результаты в список
       results.append({
           'Group A': group1,
           'Group B': group2,
           'p-value': p_value,
           'Mean A': group_data[group1].mean(),
           'Mean B': group_data[group2].mean(),
           'Absolute Lift': group_data[group2].mean()-group_data[group1].mean(),
           'Relative Lift (%)': (100*(group_data[group2].mean()/group_data[group1].mean()-1))
})

if p_value > alpha:
    print(f'pvalue={p_value} >= {alpha}')
    print('Нулевая гипотеза находит подтверждение! Нет статистически значимой разницы в ключевой метрике конверсия из регистрации в покупку')
else:
    print(f'pvalue={p_value} < {alpha}')
    print('Нулевая гипотеза не находит подтверждения! Существует статистически значимый рост ключевой метрике конверсия из регистрации в покупку')
# Создаём DataFrame из результатов
kpi_results = pd.DataFrame(results)
kpi_results
````
